```python
from transformers import pipeline
from datasets import load_dataset
import logging
import huggingface_hub
import fsspec
import numpy as np

logging.basicConfig(level=logging.DEBUG)

try:
    # print(f"transformers: {transformers.__version__}")
    # print(f"datasets: {datasets.__version__}")
    # print(f"huggingface_hub: {huggingface_hub.__version__}")
    # print(f"fsspec: {fsspec.__version__}")


    classifier = pipeline("sentiment-analysis",
                        model="distilbert-base-uncased-finetuned-sst-2-english",
                        truncation=True,
                        max_length=512)
    dataset = load_dataset("imdb",
                        split="test[:100]",
                        cache_dir="./dataset_cache")

    texts = dataset['text']
    true_labels = dataset['label']

    # Preprocessing
    texts = [text.strip()[:500] if text.strip() else "No Text" for text in texts]

    # Predictions
    results = classifier(texts[:5])
    for text, result in zip(texts[:5], results):
        print(f"Text: {text[:50]}... | Sentiment: {result['label']} ({result['score']:.2f})")

    # calculate accuracy
    predictions = classifier(texts)
    predicted_labels = [1 if res['label'] == 'POSITIVE' else 0 for res in predictions]
    accuracy = np.mean([pred == true for pred, true in zip(predicted_labels, true_labels)])
    print(f"Accuracy on 100 samples: {accuracy:.2f}")

except Exception as e:
    print(f"An error occurred: {str(e)}")
```
